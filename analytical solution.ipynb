### forced vibration anlytical solution 

##### under damped

```python
import sympy as smp
from sympy.plotting import plot
from sympy import Function, sin

x = smp.symbols("x")
a = 1.0  # value for amplitude
m = 1.0  # value for mass
k = 4.0  # value for spring constant
zeta = 0.25  # value for damping ratio
c = 2*zeta*((m*k)**0.5)
w1 = smp.sqrt(k / m)  # Natural frequency
w2 = w1 * smp.sqrt(1 - zeta ** 2)  # Damped frequency
w = 2.0  #value for external forcing frequency
t = smp.symbols("t")
p1 = smp.symbols("φ1")
p2 = smp.symbols("φ2")
p1 = 0
p2 = 0
```

```python
g = a * smp.exp(-zeta * w1 * t) * sin(w2 * t + p1)
```

```python
h = (a * sin(w * t + p2)) / ((k - m * w ** 2) ** 2 - (w * c) ** 2)
```

```python
func = Function(x)('t')
underdamped = smp.Eq(func,g+h)
```

```python
plot(g + h, (t, 0, 10), xlabel="t", ylabel="x", title="Underdamped Forced Vibration")
```

#### over damped

```python
cf = (a * smp.exp(-zeta * w2 * t*2)+2.0)* (smp.exp(-1*(w1*zeta+w2)))
```

```python
overdamped = smp.Eq(func,cf+h)
```

```python
plot(cf + h, (t, 0, 10), xlabel="t", ylabel="x", title="overdamped Forced Vibration")
```

#### critical damped

```python
c_cf = (a+2.0)*smp.exp(-w1*t)
```

```python
criticaldamped = smp.Eq(func,c_cf+h)
```

```python
plot(c_cf + h, (t, 0, 10), xlabel="t", ylabel="x", title="critical damped Forced Vibration")
```
